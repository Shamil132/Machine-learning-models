# Классификация с помщью алгоритма k ближайших соседей

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

Загрузка данных

```python
iris_dataset=load_iris()

X_train, X_test, y_train, y_test = train_test_split(iris_dataset.data, iris_dataset.target, random_state = 66)
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)

print("Правильность на тестовом наборе: {}".format(knn.score(X_test, y_test)))
```

# Выясним наиболее подходящее количество соседей

```python
iris_dataset=load_iris()
X_train, X_test, y_train, y_test = train_test_split(iris_dataset.data, iris_dataset.target, random_state = 66)

training_accuracy = []
test_accuracy = []
# пробуем n_neighbors от 1 до 10
neighbors_settings = range(1, 11)
for n_n in neighbors_settings:
# строим модель
    knn = KNeighborsClassifier(n_neighbors=n_n)
    knn.fit(X_train, y_train)
# записываем правильность на обучающем наборе
    training_accuracy.append(knn.score(X_train, y_train))
# записываем правильность на тестовом наборе
    test_accuracy.append(knn.score(X_test, y_test))

plt.plot(neighbors_settings, training_accuracy, label="правильность на обучающем наборе")
plt.plot(neighbors_settings, test_accuracy, label="правильность на тестовом наборе")
plt.ylabel("Правильность")
plt.xlabel("количество соседей")
```

Из этого графика понятно, что наиболее высокий результат наша модель показала при количестве соседей от 3 до 5
